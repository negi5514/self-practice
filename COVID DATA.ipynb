```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#importing main dataset
df = pd.read_csv('covid_19_india.csv', parse_dates=['Date'], dayfirst=True)
```

```python
df.shape
```

```python
df.head()
```

```python
# keeping only requied columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
# renaming column names
df.columns = ['date','state','cured','deaths','confirmed']
```

```python
# Looking at the earlier dates
df.head()
```

```python
# Looking at the lastest dates
df.tail()
```

```python
# current date 
today = df[df.date == '2021-08-11']
```

```python
today.head()
```

```python
# sorting data w.r.t number of confirmed dates
max_confirmed_cases=today.sort_values(by="confirmed",ascending=False)
max_confirmed_cases
```

```python
#getting states with maximum number of confirmed cases 
top_states_confirmed=max_confirmed_cases[0:5]
```

```python
top_states_confirmed
```

```python
df.head()
```

```python
 # Getting state with maximum number of confirmed cases
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="state",y="confirmed",data=top_states_confirmed,hue="state")
plt.show()
```

```python
max_death_cases=today.sort_values(by="confirmed",ascending=False)
max_death_cases
```

```python
# Getting states with maximum number deaths cases 
top_states_death=max_death_cases[0:5]
```

```python
top_states_death
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="state",y="deaths",data=top_states_death,hue="state")
plt.show()
```

```python
max_cured_cases=today.sort_values(by="cured",ascending=False)
max_cured_cases
```

```python
top_cured_cases=max_cured_cases[0:5]
```

```python
top_cured_cases
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x="state",y="cured",data=top_cured_cases,hue="state")
plt.show()
```
